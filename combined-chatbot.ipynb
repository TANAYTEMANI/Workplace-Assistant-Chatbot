```python
from langchain_community.document_loaders import PyPDFDirectoryLoader
from langchain.text_splitter import CharacterTextSplitter


DATA_PATH = "/kaggle/input/pdf-data"

def load_documents():
    document_loader = PyPDFDirectoryLoader(DATA_PATH)
    return document_loader.load()

def split_documents(documents):
    text_splitter = CharacterTextSplitter(
        chunk_size=1000,
        chunk_overlap=200,
        length_function=len,
        is_separator_regex=False,
    )
    return text_splitter.split_documents(documents)


documents = load_documents()
chunks = split_documents(documents)
#Joining the chunks in one text to pass into Pinecone Index
page_contents = [chunk.page_content for chunk in chunks]
all_content = "\n".join(page_contents)
print(all_content)
```

```python
#DOC_APP


PROMPT_TEMPLATE = """
Answer the question based only on the following context:
{context}

---
Answer the question in not more than 50 words.
Answer the question based on the above context: {question}
"""


from langchain.prompts import ChatPromptTemplate

query_text = "what is a contingency leave"

def doc_app(query_text):
    
    similar_documents = retriever.invoke(query_text)
    #similar_documents
    
    # Handle `similar_documents` assuming it is a list of `Document` objects
    context_text = "\n\n---\n\n".join([doc.page_content for doc in similar_documents])

    # Create the prompt
    prompt_template = ChatPromptTemplate.from_template(PROMPT_TEMPLATE)
    prompt = prompt_template.format(context=context_text, question=query_text)

    # Generate a response using your LLM (replace `llm` with your actual LLM instance)
    response = llm.generate(prompts=[prompt], max_tokens=50)

    # Retrieve source information from similar documents
    sources = [doc.metadata.get("source", "unknown") for doc in similar_documents]

#     # Format the response
#     formatted_response = f"Response: {response}"
#     print(formatted_response)
    
    textt = response.generations[0][0].text
    cleaned_text = textt.lstrip('\n')
    # Return the response
    return cleaned_text

doc_app(query_text)
```

```python
#TOOL_APP

from langchain_core.prompts import PromptTemplate
from langchain.chains.conversation.memory import ConversationBufferWindowMemory
from langchain.agents import initialize_agent
from langchain.agents import Tool

chat_history = []
query_text = "What is the notice period?"

def tool_app(query_text):

    json_tool = Tool(
        name="JsonTool",
        func=json_app,
        description=json_tool_desc
    )

    doc_tool = Tool(
        name="DocumentsTool",
        func=doc_app,
        description=doc_app_desc
    )

    tools = [json_tool, doc_tool]

    memory = ConversationBufferWindowMemory(
        memory_key = "chat_history",
        k=5,
        return_message = True
    )

    prompt = PromptTemplate.from_template(tool_template)
    
    conversational_agent = initialize_agent(
        agent='chat-conversational-react-description',
        tools=tools,
        llm = llm,
        prompt=prompt,
        verbose=True,
        max_iterations=2,
        early_stopping_method="generate",
        memory=memory
    )

    agent_executor = AgentExecutor.from_agent_and_tools(
        agent=conversational_agent.agent,
        tools=tools,
        verbose=True
    )

    # Define a query and update the chat history with responses
    response = agent_executor.run({"input": query_text, "chat_history": chat_history})

    # Append the new interaction to the chat history
    chat_history.append({"input": query, "response": response})

    return response

tool_app(query_text)
```
